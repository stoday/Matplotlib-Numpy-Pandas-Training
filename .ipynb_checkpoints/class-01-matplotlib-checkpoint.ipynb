# 中文顯示設定

1. Find matplotlib default font:
```
from matplotlib.font_manager import findfont, FontProperties
default_font_file = findfont(FontProperties(family=FontProperties().get_family()))
print(default_font_file)
```


2. Find the path of the setting file:
```
import matplotlib
setting_file = matplotlib.matplotlib_fname()
print(setting_file)
```


3. Edit <setting_file>, remove # from head of font.family and font.serif. And add "Microsoft JhengHei" into the Line of "font.sans-serif: xxx...."

4. Remove file ".matplotlib " in the cache. (Window: c:\User\<User_Name>, Ubuntu: /home/<User_Name>/.cache)

5. Download Chinese to default_font_file's location

6. Setting in code:
```
plt.rcParams['font.sans-serif'] = ['Microsoft JhangHei']
plt.rcParams['axes.unicode_minus'] = False
```


````python
# 畫線

X = [1, 2, 3, 4]

plt.figure()    # 建立一個圖
plt.plot(X)    # 畫出一條線
plt.grid()    # 畫格線
# plt.show()    # 秀圖
````

````python
# 畫線的樣式

X = [1, 2, 3, 4]
y = [-4, -2, 0, 2]

plt.figure()    # 建立一個圖
plt.plot(X, y, 'r-.')    # 畫出一條線
plt.grid()    # 畫格線
plt.show()    # 秀圖
````

````python
plt.figure()    # 建立一個圖
plt.plot(X, 'b-x')    # 畫出一條線
plt.grid()    # 畫格線
plt.show()    # 秀圖
````

````python
plt.figure()    # 建立一個圖
plt.plot(X, 'g-o')    # 畫出一條線
plt.grid()    # 畫格線
plt.show()    # 秀圖
````

````python
# 畫 X-Y 的線

import math

X = range(0, 1000)
y = [math.sin(k / 100) for k in X]

plt.figure()
plt.plot(X, y)
plt.grid()
plt.show()
````

````python
# 在一張圖中畫出多條線

X = range(0, 1000)
y_1 = [math.sin(k / 100) for k in X]
y_2 = [math.cos(k / 100) for k in X]

plt.figure()
plt.plot(X, y_1)
plt.plot(X, y_2)
plt.grid()
plt.show()
````

````python
# 表示多條線的說明

plt.figure()
plt.plot(X, y_1)
plt.plot(X, y_2)
plt.legend(['SIN', 'COS'])
plt.grid()
plt.show()
````

````python
# 標示標題、X 軸、Y 軸的資訊(圖例)

plt.figure()
plt.plot(X, y_1)
plt.plot(X, y_2)
plt.legend(['SIN', 'COS'])
plt.xlabel('X')
plt.ylabel('y')
plt.title('X vs. y')
plt.grid()
plt.show()
````

````python
# 指定 X 軸與 Y 軸的範圍

plt.figure()
plt.plot(X, y_1)
plt.plot(X, y_2)
plt.legend(['SIN', 'COS'])
plt.axis([0, 400, -2, 2])
plt.xlabel('X')
plt.ylabel('y')
plt.title('X vs. y')
plt.grid()
plt.show()
````

````python
# 調整圖片的大小

plt.figure(figsize=(10, 5))
plt.plot(X, y_1)
plt.plot(X, y_2)
plt.legend(['SIN', 'COS'])
plt.axis([0, 400, -2, 2])
plt.xlabel('X')
plt.xlabel('y')
plt.title('X vs. y')
plt.grid()
plt.show()
````

````python
# 畫散佈圖

import random

X_1 = [random.random() for k in range(0, 100)]
y_1 = [random.random() for k in range(0, 100)]

X_2 = [random.random() for k in range(0, 100)]
y_2 = [random.random() for k in range(0, 100)]

plt.figure(figsize=(10, 5))
plt.plot(X_1, y_1, 'b .')
plt.plot(X_2, y_2, 'rx')
plt.legend(['X_1 vs y_1', 'X_2 vs. y_1'])
plt.grid()
plt.show()
````

````python
# 畫直方圖

plt.figure(figsize=(10, 5))
plt.hist(X_1, bins=20)
plt.grid()
plt.show()
````

````python
# 多個資訊放在同一張直方圖中

# 畫直方圖

plt.figure(figsize=(10, 5))
plt.hist([X_1, X_2], bins=20)
plt.legend(['X_1', 'X_2'])
plt.grid()
plt.show()
````

````python
# 方箱圖
import numpy as np

plt.figure(figsize=(10, 5))
plt.boxplot(X_1, labels=['1st'])
plt.grid()
plt.show()
````

````python
# 方箱圖
import numpy as np

plt.figure(figsize=(10, 5))
plt.boxplot([X_1, X_2], labels=['1st', '2nd'])
plt.grid()
plt.show()
````

````python
# 佈局

fig, ax = plt.subplots(2, 2, figsize=(10, 5))
ax[0, 0].plot([1, 2, 3], 'r')
ax[0, 0].set_title('First')
ax[0, 0].grid()

ax[0, 1].plot([3, 2, 1], 'g')
ax[0, 1].set_title('Second')
ax[0, 1].grid()

ax[1, 0].plot([1, 2, 1], 'y')
ax[1, 0].set_title('Third')
ax[1, 0].grid()

ax[1, 1].plot([2, 1, 2], 'b')
ax[1, 1].set_title('4th')
ax[1, 1].grid()
````
