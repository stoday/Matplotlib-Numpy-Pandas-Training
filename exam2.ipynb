```python
from matplotlib import pyplot as plt
import numpy as np

X = np.random.randint(0, 10, size=(7, 2))

X1 = X[X>5]
X2 = np.where(X>5)

x1 = []
y1 = []

x2 = []
y2 = []

for m in range(0, X1.size):
    if X2[1][m] == 0:
        x1.append(X2[0][m])
        y1.append(X1[m])
    else:
        x2.append(X2[0][m])
        y2.append(X[X2[0][m]][0])

plt.figure()
plt.plot(X)
plt.plot(x1, y1, 'ro')
plt.plot(x2, y2, 'rx')
plt.grid()
plt.show()
```
